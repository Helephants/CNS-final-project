```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
with open('/Users/harsheeshah/Desktop/ConsoleProj/S0002.csv') as f:
    reader = csv.reader(f)
    columns_as_lists = [list(c) for c in zip(*reader)]
x1_array=[]
y1_array=[]
for i in range(1,len(columns_as_lists[18])):
    x1_array.append(float(columns_as_lists[18][i]))
    y1_array.append(float(columns_as_lists[19][i]))
xy = np.vstack([x1_array,y1_array])
z = gaussian_kde(xy)(xy)
plt.scatter(x1_array,y1_array, s=10)
print(len(x1_array))
fig, ax = plt.subplots()
ax.scatter(x1_array, y1_array, c=z, s=0.2)
plt.show()

```
